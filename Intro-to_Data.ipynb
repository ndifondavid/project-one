```python
#load the data set and print the first five rows
import pandas as pd
y=pd.read_csv(r"C:\Users\Dave\Documents\restaurant_data.csv")
print(y.head())
```

```python
# performing basic calculation
y['result']=y['Price'].mul(y['Total Customers'])
print(y['result'])
```

```python
print(y.describe())
```

```python
# DISH NAME WITH TOTAL CUSTOMERS GREATER THAN 35
Result=y[y['Total Customers']>35]['Dish Name'].value_counts()
print(Result)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.barplot(x="Dish Name", y="Total Customers", data=y, estimator=sum, ci=None,
palette="coolwarm")
plt.title("Total Customers by Dish Name")
plt.xlabel("Dish Name")
plt.ylabel("Total Customers")
plt.xticks(rotation=25)
```
